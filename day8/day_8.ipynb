```python
lst1 = []
lst2 = []
sum = 0
n_sum = 0
for i in range(5):
  num = int(input("number:"))
  if num >= 0:
    lst1.append(num)
    sum += num
  else:
    lst2.append(num)
    n_sum += num
print(lst1)
print(lst2)
print(sum)
print(n_sum)

```
