```python
def square_num(n):
  return n**2
lst1 = [1,2,3,4,5]
squared = list(map(square_num,lst1))
print("the square list is:",squared)
```

```python
celsius = [36,37,38,39]
farh = list(map(lambda x:x* 9/5 + 32,celsius))
print("celsius id:",celsius)
print("far is:",farh)
```

```python
lst2 = [1,2,3,4,5,6]
even = list(filter(lambda x:x if x % 2 == 0 else "",lst2))
print("even list is:",even)
```

```python
lst3 =  ["cricket","football","baseball","abc"]
flr = list(filter(lambda x: x if len(x) > 5 else "",lst3) )
print("filterr is:",flr)
```

```python
from functools import reduce
lst4 = [1,2,3,4,5,6]
pro = reduce(lambda x,y:x*y,lst4 )
print("product is:",pro)
```

```python
from functools import reduce
lst4 = [1,2,3,4,5,6]
maxx = reduce(lambda x,y:x if x>y else y,lst4 )
print("max is:",maxx)
```
