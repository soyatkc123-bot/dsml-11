```python
students = ["ram","Sita","Hari","Gita","Shyam","Rita"]
marks = [85,42,78,91,35,67]

grades = list(map(lambda m : "A+"if m>= 90 else ""
                            "A"if m>= 80 else ""
                            "B+"if m>=70 else ""
                            "B" if m>=60 else ""
                            "C+"if m>=50 else ""
                            "C"if m>=40 else ""
                            "NG",marks))
for name,grade in zip(students,grades):
  print(name,"-",grade)
```

```python
students = ["Ram", "Sita", "Hari", "Gita", "Shyam", "Rita"]
marks = [85, 42, 78, 91, 35, 67]

records = list(zip(students, marks))

passed_students = list(filter(lambda x: x[1] > 40, records))

print(passed_students)
```

```python
from functools import reduce
marks = [85, 42, 78, 91, 35, 67]
result = reduce(lambda x,y:x+y,marks)
print(result)
```

```python
from functools import reduce
marks = [85, 42, 78, 91, 35, 67]
results = reduce(lambda x,y:x if x>y else y,marks)
print("the highest number is:\n",results)
```
