```python
data3 = [1,2,3,4,5,6]

list(map(lambda num: num**2,data3))
```

```python
data4 = [1,2,3,4,5,6]
list(filter(lambda num:num %2 == 0 ,data4))
```

```python
from functools import reduce
lst = [1,2,3,4,5]
reduce(lambda x,y:x*y,lst)
```

```python
num = int(input("enter a number:"))
result = lambda num: "even" if num % 2==0 else "odd"
print("the result is",result(num))
```

```python
students = [
    ("ram",80),
    ("hari",95),
    ("sita",85)
]

students.sort( key=lambda x:x[1],reverse=True)
print(students)
```

```python
employees = [
    {"name":"ram","salary":50000},
    {"name":"hari","salary":30000},

]
employees.sort(key = lambda emp:emp["salary"],reverse = True)
print(employees)
```

```python
data5 = ["welcome","to","dsml","course"]
for word in data5:
  if "e" in word:
    print(word.upper())
```

```python
lstt = [1,2,3,4,5,6]
updated_num = ["even" if nums % 2 == 0 else "odd" for nums in lstt]
print(updated_num)
```

```python
list1  = [1,2]
list2 = ["a","b"]

pairs = [(x,y) for x in list1 for y in list2]

print(pairs)
```
