```python
def square(n):
  for i in range(n):
    yield i**2
a=square(3)

```

```python
print(next(a))
```

```python
def divide_numbers(num1,num2):
  try:
    result = num1/num2
    print(f"Result of {num1}/{num2}is:{result}")
  except ZeroDivisionError:
    print("error : Division by zero is not allowerd")
divide_numbers(10,5)
```

```python
def check(num):
  if num<0:
    raise ValueError("num cannot be negaive")
  print("number",num)
```

```python
check(5)
```

```python
try:
  check(-5)
except ValueError as e:
  print(e)
```

```python
def withdraw(num):
  if num<1000:
    raise ValueError("balance cannot be less than 1000")
  print("number",num)
withdraw(20)
```

```python
try:
  withdraw(200)
except ValueError as e:
  print(e)
```
