```python
class dog:
  count = 0

  def __init__(self,name,age=3):
    self.name = name
    self.age = age
    dog.count += 1

  def bark(self):
    print(f"{self.name} is barking")

  @classmethod
  def get_count(cls):
    return cls.count

  @staticmethod
  def get_vet_ingo():
    return "Dr.Saab"

d1 = dog("Puppy")
print(d1.name)
print(d1.age)
```

```python
d1.get_count()
```

```python
d1.get_vet_ingo()
```

```python
d1.bark()
```

```python
class Calculator:
  company_name = "abcd"
  def __init__(self,n1,n2,opertor):
    self.n1= n1
    self.n2 = n2
    self.operator = opertor
  def calculate(self):
    if self.operator == "+":
      return self.n1+self.n2
    elif self.operator == "-":
      return self.n1-self.n2
    elif self.operator == "/":
      return self.n1/self.n2
    elif self.operator == "*":
      return self.n1*self.n2

  @classmethod
  def show_company(cls):
    return (cls.company_name)
  @staticmethod
  def welcome():
    return "Peform an operation"

```

```python
a = int(input("enter first number:"))
b = int(input("enter second number"))
c= input("enter an operator")
c1 = Calculator(a,b,c)
```

```python
c1.calculate()
```

```python
c1.show_company()
```

```python
c1.welcome()
```
