```python
def factorial(n):
  if n == 0 or n == 1:
    return 1
  else:
    return n*factorial(n-1)
print(factorial(5))
```

```python
def fibo(n):
  if n <= 1:
    return n
  else:
    return fibo(n-1)+fibo(n-2)
number = int(input("enter a num:"))
for i in range(number):
  print(fibo(i),end=" ")
```

```python
#Nested func
def outer(x):
  def inner(y):
    return x+y
  return inner

add_=outer(4)
res = add_(5)
res
```

```python
def add(x,y):
  return x+y
def calculate(func,x,y):
  return func(x,y)
res = calculate(add,4,6)
res
```

```python
def squ(x):
  return x*x
def calc(func,x):
  return func(x)
result = calc(squ,4)
result
```

```python
#function as a value
def greeting(name):
  def hello():
    return "Hello",name
  return hello
greet=greeting("Dhiraj")
print(greet())
```

```python
def mywrapper(func):
  def myFunc():
    print("before execution")
    func()
    print("after execution")
  return myFunc
```

```python
@mywrapper
def myFunc():
  print("hello ram")
myFunc()
```

```python
def greetings(func):
  def my_name(*args,**kwargs):
    print(f"Calling{func.__name__}")
    return func(*args,**kwargs)
  return my_name

```

```python
@greetings
def my_name(name,age):
  print(f"hello{name},age{age}")
my_name("dhiraj",20)
```

```python
def greet(func):
  def say_Hello(*args,):
    print("Hellu")
    func(*args)
    print("Lala")
  return say_Hello
```

```python
@greet
def say_Hello(name,age):
  print(f"Name{name},Age:{age}")
say_Hello("Soyat",18)
```

```python
def Kangres(func):
  def jholey(*args,**kwargs):
    print("namaskar")
    func(*args,**kwargs)
    print("Jay Nepal")
  return jholey
```

```python
@Kangres
def jholey(name,age):
  print(f"Ma:{name},Umer:{age} kangres ko jholey")
jholey("Ozil",29)
```

```python
def cricsal(func):
  def wicket(*args):
    print("Play tournament")
    func(*args)
    print("Best Bowler")
  return wicket
```

```python
@cricsal
def wicket(name,sem,college):
  print(f"Name:{name},Sem{sem},College:{college}")
wicket("Swopnil",6,"Ace")
```
